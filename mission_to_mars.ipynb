### Step 1: Scraping

#### 1. Scrape the NASA Mars News Site. Collect the latest News Title and Paragraph Text.

```python
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
response = requests.get(url)

soup = bs(response.text, 'html.parser')
print(soup.prettify())
```

```python
all_articles = soup.find_all("div", class_="slide")

first_article = all_articles[0]
first_title = first_article.find("div", "content_title").text
first_paragraph =first_article.find("div", "rollover_description_inner").text

print(first_title)
print(first_paragraph)
```
